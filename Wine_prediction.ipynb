```python
sns.boxplot(data=df)
plt.xlabel('Values')
plt.title('Box Plot of All Columns')
```

```python
l1=["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]
for i in l:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upperL=q3+1.5*iqr
    lowerL=q1-1.5*iqr
    df[i]=np.where(df[i]>upperL,upperL,np.where(df[i]<lowerL,lowerL,df[i]))
```

```python
sns.boxplot(data=df)

plt.xlabel('Values')
plt.title('Box Plot of All Columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x=df.drop(columns="quality",axis=1)
x.head()
```

```python
y=df.quality
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python

x_scaled= pd.DataFrame(scale.fit_transform(x),columns =x.columns)
```

```python
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
```

```python
model1=KNeighborsClassifier()
model1.fit(x_train,y_train)
y_pred1=model1.predict(x_test)
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test,y_pred1))
accuracy_score(y_test,y_pred1)
```

```python
model2=LogisticRegression()
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
print(classification_report(y_test, y_pred2))
print(confusion_matrix(y_test,y_pred2))
accuracy_score(y_test,y_pred2)
```

```python
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
y_pred3=model3.predict(x_test)
print(classification_report(y_test, y_pred3))
print(confusion_matrix(y_test,y_pred3))
accuracy_score(y_test,y_pred3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model4=RandomForestClassifier()
model4.fit(x_train,y_train)
y_pred4=model4.predict(x_test)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test,y_pred4))
accuracy_score(y_test,y_pred4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
model5=LinearRegression()
model5.fit(x_train,y_train)
y_pred5=model5.predict(x_test)

metrics.r2_score(y_test,y_pred5)
```
